> The column below indicates the degree of anxiety, and the higher the number, the higher the stability. Based on this column, I would like to examine the correlation between a total of five extraversion scales and psychological stability.

#### 1. Correlation between extraversion scale 1 and stability
> The report examines the relationship between psychological stability and participation in social events, which is one of the measures of extraversion.
The column title can be checked through the "pt.final" data frame by using "head()" function. First, create a linear regression model by using the "lm()" function. The stability, which is the dependent variable to be predicted, is set as "Stability" on the y=axis, and participation in social events, an independent variable used for prediction, is designated as "Enjoy social events" on the x-axis. Designate the model created in this way as an object called "fit1". The model can be evaluated by using "summary()" function.

> It can be predicted that the stability value increases by 0.2988 for every increase in the degree of enjoying social events by 1. They have a positive correlation, and the equation of the linear model that explains the relationship between the two variables is as follows.<br>
Y = 3.4002 + 0.2988*X

> A linear regression model can be visualized as graph with the "plot()" function. Set the graph title and the titles for each of the x-axis and y-axis. Set the font size to 1.5 using "cex.lab". A regression line is indicated using "abline()", and its color is red.


```r
head(pt.final$`X`,1)
```

```r
fit1 <- lm(ptnew$X.50 ~ ptnew$X, data=ptnew)
summary(fit1)
p1 <- plot(ptnew$X, ptnew$x.50, main="Correlation between extraversion scale 1 and stability", xlab = "Enjoy social events ", ylab = "Stability", cex.lab="1.5"); abline(fit1, col="red")
```

## **5) Multiple Linear Regression**

#### 1. Correlation between introversion scales and stability
> Check the column names through the "pt.final" data frame and set them as independent variables. <br>X14 represents "Be farther away from the social event" and X23 represents the degree of preference for having conversation with familiar people in unfamiliar places. X36 indicates the degree to which one prefers a quiter, less crowded place.

```r
head(pt.final$`X.14`,1)
head(pt.final$`X.23`,1)
head(pt.final$`X.36`,1)
```

> Multiple linear regression predicts by inputting two or more multiple variables as variables used for prediction. The formula is the same as simple linear regression model as "lm(dependent variable ~ independent variable)", and several independent variables are connected with a '+' sign.
Select the introversion scales as opposed to the extraversion scale and compare it with stability. The intercept is 8.05932. Based on this, the regression equation is as follows.<br>
Stability(Y) = (-0.33861)*X14 + (-0.11051)*X23 + (-0.01245)*X36 + 8.05932<br>
<br>
With X14 and X23 under control, if X36 increases by 1, stability decreases by 0.01245. In the state where X14 and X36 are controlled, when X23 increases by 1, stability decreases by 0.11051. With X23 and X36 are controlled, if X14 increases by 1, the stability decreases by 0.33861.<br>
<br>
The multiple r-squared is 19.67% and the adjussted r-squared is 12.37%. This is higher than the rest of the simple linear regressions, except for the extraversion scale 3, which had the highest explanatory power among the simple linear regressions.

```r
model1 <- lm(ptnew$X.50 ~ ptnew$X.14 + ptnew$X.23 + ptnew$X.36, data = ptnew)
summary(model1) 
```

#### 2. Correlation between regret scales and stability
> Check the column names through the "pt.final" data frame and set them as independent variables. <br>X3 represents the degree of regret about what have said and X13 represents the degree to which one keeps thinking about mistakes. X36 shows how much you doubt yourself if you make a mistake.

```r
head(pt.final$`X.3`,1)
head(pt.final$`X.13`,1)
head(pt.final$`X.22`,1)
```

> Next, I would like to look at the correlation between an index indicating the degree of regret and stability. Th intercept in here is 10.36918. The regression equation is as follows.<br>
Stability(Y) = (-0.57342)*X3 + (-0.19126)*X13 + 0.02957*X22 + 10.36918<br>
<br>
When X3 and X13 are controlled and X22 is increased by 1, stability increases by 0.02957. With X3a and X22 under control, if X13 increases by 1, stability decreases by 0.19126. With X13 and X22 under control, if X3 increases by 1, stability decreases by 0.57342.<br>
<br>
The multiple r-squared is 31.13% and the adjussted r-squared is 24.86%. It is shown that the model2 has a higher explanatory power than model1. However, it is still lower than the explanatory power of the extraversion scale 3, and it can be seen that the extraversion scale 3 model still shows higher accuracy.


```r
model2 <- lm(ptnew$X.50 ~ ptnew$X.3 + ptnew$X.13 + ptnew$X.22, data = ptnew)
summary(model2)
```
